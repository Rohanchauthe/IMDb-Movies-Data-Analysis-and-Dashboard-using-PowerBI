```python
import pandas as pd
import requests
import time

API_KEY = "a9622c0c"   # replace with your OMDb API key
INPUT_FILE = "IMDB-Movie-Data.csv"
OUTPUT_FILE = "IMDB-Movie-Data-With-Posters.csv"

df = pd.read_csv(INPUT_FILE)
poster_urls = []

print(f"Loaded {len(df)} movies.")
print("Fetching poster URLs...\n")

for index, row in df.iterrows():
    title = row["Title"]
    year = row["Year"]

    enc_title = title.replace(" ", "+")
    url = f"http://www.omdbapi.com/?t={enc_title}&y={year}&apikey={API_KEY}"

    try:
        response = requests.get(url, timeout=10).json()
        poster = response.get("Poster", None)
    except Exception as e:
        print(f"[{index}] ERROR fetching {title}: {e}")
        poster = None

    poster_urls.append(poster)
    print(f"[{index+1}/{len(df)}] {title} → {poster}")

    time.sleep(0.20)   # small delay for API limits

df["PosterURL"] = poster_urls
df.to_csv(OUTPUT_FILE, index=False)

print("\nDone!")
print(f"Poster-enriched CSV saved as: {OUTPUT_FILE}")
```
